# Exploratory Data Analysis (EDA) for Grant Application Success Prediction

In this notebook, we will explore the dataset to understand its structure, identify patterns, and gather insights that will inform subsequent modeling decisions.

## Step 1: Load the Preprocessed Data

First, we load the preprocessed training and test datasets.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the preprocessed training and test datasets
train_df = pd.read_csv('data/preprocessed_train.csv')
test_df = pd.read_csv('data/preprocessed_test.csv')

# Convert inf values to NaN
train_df.replace([np.inf, -np.inf], np.nan, inplace=True)
test_df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Display the first few rows of the training data
print("Training Data:")
print(train_df.head())
print("\nTest Data:")
print(test_df.head())
```

## Step 2: Understand the Data Structure and Summary Statistics

We will examine the structure and summary statistics of the dataset to get an overview of the data.


```python
# Display the data types and non-null values
print("Data Types and Non-null Values:")
print(train_df.info())

# Display summary statistics for numerical columns
print("\nSummary Statistics:")
print(train_df.describe())
```

## Step 3: Visualize the Distribution of Key Variables

We will visualize the distribution of key numerical and categorical variables to understand their distributions and identify any potential anomalies.


```python
# Distribution of the target variable 'Grant.Status'
plt.figure(figsize=(8, 6))
sns.countplot(x='Grant.Status', data=train_df)
plt.title('Distribution of Grant Status')
plt.show()

# Histograms for numerical features
numerical_cols = ['RFCD.Percentage.1', 'SEO.Percentage.1', 'Year.of.Birth.1', 
                  'Number.of.Successful.Grant.1', 'Number.of.Unsuccessful.Grant.1']

for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(train_df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

## Step 4: Examine Relationships Between Features and the Target Variable

We will explore the relationships between key features and the target variable 'Grant.Status' to identify any significant patterns.


```python
# Box plots for numerical features vs. Grant.Status
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Grant.Status', y=col, data=train_df)
    plt.title(f'{col} vs. Grant Status')
    plt.show()

# Count plots for categorical features vs. Grant.Status
categorical_cols = [ 'Grant.Category.Code', 'Contract.Value.Band...see.note.A', 
                    'RFCD.Code.1', 'SEO.Code.1', 'Person.ID.1', 'Role.1']

for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=col, hue='Grant.Status', data=train_df)
    plt.title(f'{col} vs. Grant Status')
    plt.xticks(rotation=90)
    plt.show()
```
